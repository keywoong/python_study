### 2. 리스트 정렬하기


#### 얕은 복사

```python
unordered_list = [5,4,6,1,8,3,5,8,1] # 정렬되지 않은 리스트
ordered_list = unordered_list
ordered_list.sort()
print(ordered_list, unordered_list)
```

```python
print('unordered_list의 주소값 : ', id(unordered_list))
print('ordered_list의 주소값 : ', id(ordered_list))
print(id(unordered_list) == id(ordered_list))
```

복사를 할 때 값만 복사하는 것이 아니라 주소값까지 복사하므로 unordered_list나 ordered_list가 바뀌면 다른 것도 영향을 받는다는 것을 알 수 있다.

#### 얕은 복사2

```python
unordered_list = [5,4,6,1,8,3,5,8,1] # 정렬되지 않은 리스트
ordered_list = unordered_list.copy()
ordered_list.sort()
print(ordered_list, unordered_list)
```

```python
print('unordered_list의 주소값 : ', id(unordered_list))
print('ordered_list의 주소값 : ', id(ordered_list))
print(id(unordered_list) == id(ordered_list))
```

이렇게 하면 서로 영향을 주지 않는다.

#### 깊은 복사

```python
import copy

unordered_list = [5,4,6,1,8,3,5,8,1] # 정렬되지 않은 리스트
ordered_list = copy.deepcopy(unordered_list)
ordered_list.sort()
print(ordered_list, unordered_list)
```

```python
print('unordered_list의 주소값 : ', id(unordered_list))
print('ordered_list의 주소값 : ', id(ordered_list))
print(id(unordered_list) == id(ordered_list))
```

### 3. Passing Arguments

- 함수에 입력되는 arguments는 다양한 형태를 가진다 <br>
(parameter는 문자, argument는 숫자, 값을 의미)<br>

- 1) Keyword arguments
- 2) Default arguments
- 3) Variable-length arguments

#### 3.1 Keyword Arguments
- 함수에 입력되는 paramter의 변수명을 사용. arguments를 넘김

```python
def print_something(my_name, your_name):
    print("Hello {0}, My name is {1}".format(your_name, my_name))
print_something("David", "Tom")
print_something(your_name = "Tom", my_name = "David")
```

#### 3.2 Default Arguments
- parameter의 기본값을 사용. 입력하지 않을 경우 기본값을 출력한다.

```python
def print_something2(my_name, your_name = "Kate"):
    print("Hello {0}, My name is {1}".format(your_name, my_name))
print_something2("David")
print_something2("David", "Jay")
```

#### 3.3 Variable-length Arguments 가변인자 (*args)

- 개수가 정해지지 않은 변수를 함수의 parameter로 사용한다.
- Keyword arguments와 함께, argument 추가가 가능
- Asterisk 기호를 사용하여 함수의 parameter를 표시함.
- 입력된 값은 <b>tuple type</b>으로 사용할 수 있음
- 가변인자는 오직 <b>한 개</b>만, <b>맨 마지막 parameter위치</b>에 사용 가능.

```python
# 가변인자 
def asterisk_test1(a,b,*args):
    return a+b+sum(args)
print(asterisk_test1(1,2,3,4,5))
```

```python
def asterisk_test2(a,b,*args):
    return args
print(asterisk_test2(1,2,3,4,5)) # 튜플
```

```python
def asterisk_test3(a,b,*args):
    return args[0]
print(asterisk_test3(1,2,3,4,5))
```

```python
def asterisk_test_unpacking(*args):
    x,y,z = args
    return x,y,z
asterisk_test_unpacking(1,2,3)
```

#### 3.4 Keyword variable-length 키워드 가변인자 (**kwargs)
- Parameter 이름을 따로 지정하지 않고 입력하는 방법
- Asterisk 2개를 사용하여 함수의 Parameter를 표시한다.
- 입력된 값은 <b>dict type</b>으로 사용할 수 있음.
- 키워드 가변 인자는 오직 한개만, 기존 가변인자 다음에 사용한다.

```python
def kwargs_test_1(**kwargs):
    print(kwargs)
def kwargs_test_2(**kwargs):
    print(kwargs)
    print("First value is {first}".format(**kwargs))
    print("Second value is {second}".format(**kwargs))
    print("Third value is {third}".format(**kwargs))
```

```python
kwargs_test_1(first=3, second=4, third=5)
```

```python
kwargs_test_1(one=3, two=4, three=5)
```

```python
kwargs_test_2(first = 10, second = 20, third = 30)
```

```python
kwargs_test_2(first = 10, second = 20, third = 30, fourth = 40)
```

```python
kwargs_test_2(one = 10, second = 20, third = 30)
```

#### 3.5 모든 arguments(키워드, 기본, 가변인자, 키워드가변인자)를 사용한 예

```python
def kwargs_test_3(one, two=2, *args, **kwargs):
    print("One is {}".format(one))
    print("Two is {}".format(two))
    print("args is {}".format(args))
    print("kwargs is {}".format(kwargs))
```

```python
kwargs_test_3(1,2,3,"hello",5,6,7, first=3,second=4,third=5)
```

### 5. 선형대수

$x_1 + 2x_2 + 3x_3 = 4$<br>
$2x_1 + 5x_2 + 3x_3 = 5$<br>
$x_1 + 8x_3 = 9$<br>

```python
import numpy as np
matrix_a = np.array([
    [1,2,3],
    [2,5,3],
    [1,0,8]
])

matrix_b = np.array([
    [4],[5],[9]
])

matrix_a_inverse = np.linalg.inv(matrix_a)

x = np.dot(matrix_a_inverse, matrix_b)
x
```

$x_1 + 2x_2 + 3x_3 = 1$<br>
$2x_1 + 5x_2 + 3x_3 = 6$<br>
$x_1 + 8x_3 = -6$<br>

```python
matrix_a = np.array([
    [1,2,3],
    [2,5,3],
    [1,0,8]
])

matrix_b = np.array([
    [1],[6],[-6]
])

matrix_inverse_a = np.linalg.inv(matrix_a)

x = np.dot(matrix_inverse_a, matrix_b)
x
```

### 6. Numpy를 이용한 동력계 문제 해결

```python
def x(t):
    matrix_a = np.array([
        [0.8,0.1],
        [0.2,0.9]
    ])
    
    init_vector = np.array([0.5,0.5])
    
    if t == 0:
        return init_vector
    else:
        result = init_vector
        for i in range(t):
            result = matrix_a.dot(result)
    return result
```

```python
print('초기의 각 채널의 점유율 : ', x(0))
print('1년 후 각 채널의 점유율 : ', x(1))
print('2년 후 각 채널의 점유율 : ', x(2))
print('5년 후 각 채널의 점유율 : ', x(5))
print('100년 후 각 채널의 점유율 : ', x(100))
```

### 8. 각 Unit 별로 남은 Cycle횟수를 알고자 한다. RemainingCycle이라는 Column을 추가하고, 나은 Cycle횟수를 나타내어라

```python
data.head()
```

1. UnitNumber를 기준으로 groupby하고,
2. 각 UnitNumber들의 Cycle의 최대값을 column으로 만들어준다.

```python
max_cycle = data.groupby('UnitNumber')['Cycle'].max().reset_index()
max_cycle.rename(columns = {'Cycle':'MaxCycle'}, inplace = True)
max_cycle
```

```python
temp_df = pd.merge(data, max_cycle, on='UnitNumber',how = 'left')

temp_df
```

```python
data['RemainingCycle'] = temp_df['MaxCycle'] - data['Cycle']
data.head()
```

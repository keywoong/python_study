## 1. Regression

### 1.1 Regression은 Supervised Learning
기계학습은 크게 3가지로 구성된다.
- Supervised Learning : 레이블된 데이터로 학습, 미래 데이터 예측.<br>
ex) Regression, Classification <br>
- Unsupervised Learning : 레이블 없이 학습, 데이터의 숨겨진 구조 및 특징 발견<br>
ex) Clustering
- Reinforcement Learning : 보상 시스템으로 학습, 의사결정을 위한 최적의 액션 선택
<br>

본 편에서는 Supervised Learning의 일종인 Regression을 다룬다.

Regression이란 데이터와 오차가 적은 선을 찾는 것을 말함.

### 1.2 Type of Regression
1. Linear Regression, 선형회귀
- 변수들이 서로 선형적으로 연결되어 있는 경우에 사용

2. Ridge Regression, 릿지 회귀
- 독립 변수들 사이에 높은 상관관계가 있을 경우 사용
- Regularization 기법으로 모델의 복잡성을 줄임

3. Lasso Regression, 라쏘 회귀
- 릿지 회귀와 같이 모델의 복잡성을 줄임
- 회귀 계수의 절대 사이즈를 규제함으로써 복잡성을 줄임

4. Polynomial Regression, 다항 회귀
- 선형 모델을 사용하여 비선형 데이터 집합을 모델링
- 비선형 곡선을 사용함.

## 2. Cost Function


데이터들을 잘 표현하는 직선을 구하기 위해선 함수의 예측값과 실제 값의 오차를 구해야 함.<br>
1) 오차의 합<br>
$(\hat{y}^{(1)} - y^{(1)}) + (\hat{y}^{(2)} - y^{(2)}) + (\hat{y}^{(3)} - y^{(3)}) + (\hat{y}^{(4)} - y^{(4)})$
<br><br>
그런데 오차값은 양수 또는 음수가 될 수 있기 때문에 오차를 합치면 상쇄가 될 수 있다.

2) 제곱의 합으로 변환<br>
$(\hat{y}^{(1)} - y^{(1)})^2 + (\hat{y}^{(2)} - y^{(2)})^2 + (\hat{y}^{(3)} - y^{(3)})^2 + (\hat{y}^{(4)} - y^{(4)})^2$
<br><br>
= $\Sigma_{i=1}^{n} (\hat{y}^{(i)} - y^{(i)})^2 $

위의 식을 통해 함수의 예측값과 실제값의 오차를 알 수 있다. 
<br>
따라서 이를 
- 오차함수 (Error Function)
- 손실함수 (Loss Function)
- 비용함수 (Cost Function)이라고 부른다.

## 3. Gradient Descent

- 미분계수를 통해 현재 parameter가 Cost function에 어떤 영향을 미치는지 확인
- 미분계수가 0이 되는 방향으로 업데이트하기

```python
# x^2의 최소값 찾기
x = range(-10,11)
y = [value**2 for value in x]

plt.plot(x,y) # 연속적인 값이므로 plot을 사용한다.
plt.show() # 그래프를 보여준다.
```

```python
x = range(-10,11)
y = [value**2 for value in x]

plt.plot(x,y)

x = 10
derivate = []
y = []
learning_rate = 0.01
epochs = 1000

for i in range(epochs):
    old_value = x
    y.append(old_value**2)
    derivate.append(old_value - learning_rate*2*old_value)
    x = old_value - learning_rate*2*old_value
    
plt.scatter(derivate,y)
plt.show()
```

여기서 총 2개의 hyperparameter가 존재한다.
1. Learning Rate를 몇으로 할 것인가?
2. 얼마나 많은 epochs을 설정할 것인가?

### 3.1 Learing Rate를 몇으로 할 것인가?

```python
# learning rate = 0.9
x = range(-10,11)
y = [value**2 for value in x]

plt.plot(x,y)

x = 10
derivate = []
y = []
learning_rate = 0.9
epochs = 1000

for i in range(epochs):
    old_value = x
    y.append(old_value**2)
    derivate.append(old_value - learning_rate*2*old_value)
    x = old_value - learning_rate*2*old_value
    
plt.scatter(derivate,y)
plt.show()
```

learning rate가 너무 크다.

```python
# learning rate = 0.0001
x = range(-10,11)
y = [value**2 for value in x]

plt.plot(x,y)

x = 10
derivate = []
y = []
learning_rate = 0.0001
epochs = 1000

for i in range(epochs):
    old_value = x
    y.append(old_value**2)
    derivate.append(old_value - learning_rate*2*old_value)
    x = old_value - learning_rate*2*old_value
    
plt.scatter(derivate,y)
plt.show()
```

learning rate가 너무 작다.

## 4. Linear Regression with GD
y = 4x+6을 예측하는 모델

```python
np.random.seed(0) #랜덤시드 고정

x = 2*np.random.rand(100,1)
# random.rand : 0~1 까지의 표준정규분포 난수 array를 반환
y = 6+4*x+np.random.randn(100,1)
# random.randn : 평균0, 표준편차1의 가우시안 표준정규분포 난수 array 반환

plt.figure(figsize = (5,5))
plt.scatter(x,y)
plt.show()
```

```python
# get_cost
def get_cost(y,y_pred):
    N = len(y)
    cost = np.sum(np.square(y-y_pred))/N
    return cost
```

```python
# weight_update
def get_weight_update(w,b,x,y,learning_rate = 0.01):
    N = len(y)
    
    #크기가 각각 w,b와 동일한 크기를 가지고 0값으로 초기화된 updated array
    w_update = np.zeros_like(w)
    b_update = np.zeros_like(b)
    
    y_pred = np.dot(x,w.T) + b
    diff = y-y_pred
    
    b_factors = np.ones((N,1))
    
    # w,b를 업데이트할 값 계산
    # cost function을 미분한 식
    w_update = -(2/N)*learning_rate*(np.dot(x.T, diff))
    b_update = -(2/N)*learning_rate*(np.dot(b_factors.T, diff))
    
    return w_update, b_update
```

```python
def gradient_descent_steps(x,y,epochs):
    # w,b 를 모두 0으로 초기화
    w = np.ones((1,1))
    b = np.ones((1,1))
    
    for _ in range(epochs):
        w_update, b_update = get_weight_update(w,b,x,y,learning_rate = 0.01)
        w = w-w_update
        b = b-b_update
    return w,b
```

```python
w,b = gradient_descent_steps(x,y,epochs = 1000)
print(w,b)
y_pred = w*x + b
print(get_cost(y,y_pred))
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x,y)
plt.plot(x,y_pred, c = 'r')
plt.show()
```

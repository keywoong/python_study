Column 연산으로 새로운 Column 생성하기

```python
# Column 연산으로 총 인구수 구하기
data['인구 (명)_남+여'] = data['인구 (명)_남'] + data['인구 (명)_여']
data
```

```python
# Column 연산으로 남녀 인구수 차이 구하기
data['인구 (명)_남-여'] = data['인구 (명)_남'] - data['인구 (명)_여']
data
```

```python
# Column 연산으로 남녀 비율 구하기

data['남자/인구수'] = data['인구 (명)_남']/data['인구 (명)_남+여']
data['여자/인구수'] = data['인구 (명)_여']/data['인구 (명)_남+여']
data
```

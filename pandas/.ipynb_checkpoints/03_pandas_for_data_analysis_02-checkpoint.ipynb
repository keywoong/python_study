#### txt파일 불러오기 (`read_csv()`)
`read_csv()`로 CMAPSS.txt 파일을 불러온다.

```python
import pandas as pd

data = pd.read_csv('CMAPSS.txt', sep = ' ', header = None, encoding = 'cp949')
data.head()
```

- `sep = ' '`는 구분자를 ' '로 한다.
- `header = None`은 Column을 설정하지 않는다.

#### 칼럼 지우기 (`drop()`)
이제 `drop()`함수를 통해 <u>NaN값으로 도배된 Column 26,27</u> 을 지운다.

```python
data.drop([26,27], axis = 1, inplace = True)
data.head()
```

- `axis = 1` : Column 방향으로 drop (0이면 row방향)
- `inplace = False` : data에 함수를 적용하지만 반환만 될 뿐 data자체가 바뀌지 않음, 반대로 `inplace = True`인 경우에는 data 자체가 변경됨

#### 칼럼명 지정
숫자로 되어 있는 칼럼명에 이름을 넣어준다.

```python
index_columns = ['UnitNumber', 'Cycle']
op_setting_columns = ['Op_Setting_' + str(i) for i in range(1,4)]
sensor_columns = ['Sensor_' + str(i) for i in range(1,22)]

columns = index_columns + op_setting_columns + sensor_columns
data.columns = columns
data.head()
```

#### 특정 칼럼의 요약정보 보기

```python
data[data['UnitNumber'] == 77]
```

```python
# describe() 함수를 이용하면 DataFrame의 요약정보를 알 수 있다.
data[data['UnitNumber'] == 77].describe()
```

#### 특정 칼럼 기준으로 다른 칼럼들 보기 (`groupby()`)
각 cycle 별로 Sensor_1의 평균을 보고 싶을 때<br>
Cycle별로 groupby시키고, 그 안에 있는 Sensor1을 불러와 평균을 구함.

```python
# groupby() 함수를 이용하여 특정 Column 기준으로 묶을 수 있음.
data.groupby('Cycle')['Sensor_1'].mean()
```

[예제] 각 Unit 별로 남은 Cycle 횟수를 알고자 한다. RemainingCycle이라는 Column을 추가하고, 남은 Cycle횟수를 나타내라.

```python
max_cycle = data.groupby('UnitNumber')['Cycle'].max().reset_index() # max_cycle을 DataFrame으로 만들기 위해 reset_index() 사용
max_cycle.rename(columns = {'Cycle' : 'MaxCycle'}, inplace = True)

max_cycle
```

```python
# pd.merge()함수를 이용해 DataFrame을 합친다.
temp_df = pd.merge(data, max_cycle, on = 'UnitNumber', how = 'left')
# temp란 임의로 만든 변수
# merge() 속 on은 기준이 되는 Column을 의미
# merge() 속 how는 기준이 되는 DataFrame을 의미
data['RemainingCycle'] = temp_df['MaxCycle'] - data['Cycle']
data.head()
```

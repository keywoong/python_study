## 1. Data Structure
- 메모리상에 데이터를 효율적으로 관리하는 방법

### 1. Tuple
- 값의 변경이 <b>불가능</b>한 리스트
- 선언시 ()를 사용
- 리스트의 연산, 인덱싱, 슬라이싱 등을 동일하게 사용

```python
t = (1,2,3)
print(t+t)
print(t*2)
print(len(t))
print(type(t))
```

```python
t[1] = 5#변경이 불가능
```

값이 하나이 tuple은 반드시 ','를 붙여야 한다. 아니면 tuple이 아닌 정수로 인식함.

```python
t1 = (1)
print(type(t1))
t2 = (1,)
print(type(t2))
```

tuple은 학번, 이름, 우편번호 등 프로그램을 작동하는 동안 변경되지 않아야 하는 데이터를 저장할 때 사용한다.

### 1.2 Set
- 값을 순서없이 저장
- 중복을 불허하는 자료형
- Set 객체 선언을 이용하여 객체 생성

```python
s = set([1,2,3,1,2,3])
print(s)#중복을 불허함.
s.add(1)
print(s)
s.remove(1)
print(s)
s.update([1,2,3,4,5])
print(s)
s.clear()
print(s)
```

set의 연산.<br>
수학에서 활용하는 다양한 집합 연산 가능

```python
s1 = set([1,2,3,4,5])
s2 = set([3,4,5,6,7])

#합집합
print(s1.union(s2))
print(s1|s2)
#교집합
print(s1.intersection(s2))
print(s1&s2)
#차집합
print(s1.difference(s2))
print(s1-s2)
```

#### 1.4.1 deque
- stack과 queue를 지원하는 모듈
- List에 비해 효율적인 자료 저장 방식을 지원함.
<br>

<b>stack</b>
- 후입선출, 가장 마지막에 삽입된 자료가 가장 먼저 삭제된다.
- 끝부분이 막혀있는 형태<br>

<b>queue</b>
- 선입선출, 먼저 삽입된 자료가 먼저 삭제된다.
- 약간 밑빠진 독같은 느낌.

#### stack, 후입선출
끝부분이 막혀있다.

```python
from collections import deque
#stack 생성
deque_list = deque()
for i in range(5):
  deque_list.append(i)
  print(deque_list)
```

```python
#stack 제거
for _ in range(5):
  deque_list.pop()
  print(deque_list)
```

stack은 후입선출이라 가장 나중에 들어온 것부터 빠져나가는 것을 볼 수 있다.

#### queue, 선입선출
밑빠진 독, 먼저 들어온 것이 먼저 나간다.

```python
from collections import deque
#queue 생성
deque_list = deque()
for i in range(5):
  deque_list.append(i)
  print(deque_list)
```

```python
#queue 제거
for _ in range(5):
  deque_list.popleft()
  print(deque_list)
```


- deque는 rotate, reverse 등 Linked List의 특성을 지원한다.
- 기존 List 형태의 함수를 모두지원한다.

```python
#deque_list 생성
deque_list = deque()
for i in range(5):
  deque_list.append(i)
print(deque_list)
```

```python
deque_list.rotate(1)
print(deque_list)
#rotate(i) : 0번째가 i에 가있도록
```

```python
deque_list.rotate(1)
print(deque_list)
```

```python
deque_list.rotate(2)
print(deque_list)
```

```python
print(deque_list)
print(deque(reversed(deque_list)))
```

```python
print(deque_list)
deque_list.extend([5,6,7])
print(deque_list)
```

```python
print(deque_list)
deque_list.extendleft([5,6,7])
print(deque_list)
```

deque와 general list와의 시간 차이<br>
- 기존 list보다 효율적인 자료구조를 제공
- 효율적 메모리 구조로 처리 속도 향상

```python
from collections import deque
import time

start_time = time.process_time()
deque_list = deque()
for _ in range(10000):
  for i in range(10000):
    deque_list.append(i)
    deque_list.pop()
print(time.process_time()-start_time, 'seconds')
```

```python
import time

start_time = time.process_time()
general_list = []
for _ in range(10000):
  for i in range(10000):
    general_list.append(i)
    general_list.pop()
print(time.process_time()-start_time,'seconds')
```

#### 1.4.2 default dict

```python
d = dict()
print(d['first'])
```

#### iterator
- 반복가능한 객체, 반복문을 활용해서 데이터를 순회하면서 처리하는 것을 의미함.
- List, Dictionary, Set, Tuple, String 등 순회 가능한 Sequence 객체들에 대해서 iterable object라고 지칭함.

```python
tuple1 = (1,2,3)
iterator1 = iter(tuple1)
print(iterator1)
for _ in range(4):
  print(iterator1.__next__())
```

### 1.4.1 Namedtuple
- Tuple 형태로 Data 구조체로 저장하는 방법
- 저장하는 Data의 variable을 사전에 지정해서 저장함.
- Tuple은 인덱스를 통해 데이터에 접근은 가능한데 NamedTuple은 인덱스도 가능하고 추가로 key값으로도 접근이 가능.

```python
from collections import namedtuple
Point = namedtuple('Point',['x','y'])
p = Point(11,y=22)
print(p)
```

```python
print(p[0]+ p[1])
```

```python
x,y, = p
print(x,y)
```

```python
print(p.x + p.y)
```

```python
print(Point(x=11, y=22))
```

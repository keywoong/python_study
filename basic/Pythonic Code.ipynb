## 1. Pythonic Code

```python
# 여러 단어들을 하나로 붙일 때 1
colors = ['red','blue','green','yellow']
results = ''
for color in colors:
  results += color
print(results)
```

```python
# 여러 단어들을 하나로 붙일 때 2
colors = ['red','blue','green','yellow']
results = ''.join(colors)
print(results)
```

#### List Comprehension
- 기존 List를 사용하여 간단히 다른 List를 만드는 기법
- 포괄적인 List, 포함되는 List라는 의미
- 파이썬에서 가장 많이 사용되는 기법 중 하나
- 일반적으로 for+append 보다 속도가 빠름

ex) 0부터 9까지 숫자를 result라는 리스트로 만들고 싶음.

```python
result1 = []
for i in range(10):
  result1.append(i)
print('append 결과', result1)

result2 = [i for i in range(10)]
print('list comprehension 결과',result2)
```

ex) 0부터 9까지 숫자 중 짝수로만 result라는 리스트로 만들고 싶음.

```python
result1 = []
for i in range(10):
  if i %2==0:
    result1.append(i)
print(result1)

result2 = [i for i in range(10) if i%2==0]
print(result2)
```

ex) Hello와 World를 글자별로 합쳐서 result에 넣고 싶음.

```python
# append
word_1 , word_2 = 'Hello', 'World'
result1 = []
for i in word_1:
  for j in word_2:
    result1.append(i+j)
print(result1)
```

```python
word_1, word_2 = 'Hello', 'World'
result2 = [i+j for i in word_1 for j in word_2]
print(result2)
```

ex) case_1의 요소와 case_2의 요소를 합쳐서 result에 넣어주는데, 만약 두 요소가 같으면 추가하지 않음.

```python
case_1 = ['A','B','C']
case_2 = ['D','E','A']
result = []
for i in case_1:
  for j in case_2:
    if not i==j:
      result.append(i+j)
print(result)
```

```python
case_1 = ['A','B','C']
case_2 = ['D','E','A']
result = [i+j for i in case_1 for j in case_2 if i!=j]
print(result)
```

다차원 list

```python
case_1 = ['A','B','C']
case_2 = ['D','E','A']

result1 = [i+j for i in case_1 for j in case_2]
print(result1)

result2 = [[i+j for i in case_1] for j in case_2]
print(result2)
```

zip<br>
두 개의 list의 값을 병렬적으로 추출함

```python
alist = ['a1','a2','a3']
blist = ['b1','b2','b3']
for a,b in zip(alist, blist):
  print(a,b) #병렬적으로 추출함.
```

```python
a,b,c = zip(
    (1,2,3),
    (10,20,30),
    (100,200,300)
)
print(a,b,c)
```

```python
[sum(x) for x in zip(
    (1,2,3), (10,20,30), (100,200,300)
)]
```

lambda<br>
- 함수 이름없이 함수처럼 쓸 수 있는 익명함수
- 수학의 람다 대수에서 유래함
- 코드의 직관성이 떨어져서 권장하지는 않지만, legacy library에서 사용되었음.

```python
#general funciton
def f(x,y):
  return x+y
print(f(1,2))
```

```python
#lambda function
f = lambda x,y :x+y
print(f(1,2))
```

```python
f = lambda x:x**2
print(f(3))
```

```python
f = lambda x:x/2
print(f(4))
```

```python
print((lambda x:x+1)(5))
```

map function<br>
Sequence 자료형 각 element에 동일한 function을 적용함.

```python
ex = [1,2,3,4,5]
f = lambda x:x**2
print(list(map(f,ex)))
```

```python
ex = [1,2,3,4,5]
f = lambda x,y : x+y
print(list(map(f, ex,ex)))
```

asterick<br>
단순곱셈, 제곱연산, 가변 인자 활용 등 다양하게 사용됨.

```python
# *args
def asterisk_test(a,*args):
  print(a,args)
  print(type(args))
asterisk_test(1,2,3,4,5,6)
```

```python
# **kwargs
def asterisk_test(a,**kwargs):
  print(a, kwargs)
  print(type(kwargs))
asterisk_test(1, b=2, c=3, d=4, e=5, f=6)
```

```python
def asterisk_test(a,*args):
  print(a,args)
  print(a,*args)
  print(type(args))
asterisk_test(1,2,3,4,5,6)
```

```python
asterisk_test(1,*(2,3,4,5,6))
```

```python
asterisk_test(1,2,(2,3,4,5))
```
